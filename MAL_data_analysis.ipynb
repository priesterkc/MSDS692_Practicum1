```python
ml_anime_df = repack_anime_df[['mean', 'num_list_users', 'num_favorites', 'media_type', 
 'rating','studios', 'genres', 'demographics']].copy()

ml_anime_df.head()
```

```python
ml_anime_df.info()
```

```python
# drop null from genres and ratings

len(ml_anime_df.dropna())
```

```python
ml_anime_df.dropna(inplace=True)
```

```python
# remove shows that have a mean of zero (these haven't aired yet or haven't been rated by users yet)

ml_anime_df = ml_anime_df.loc[ml_anime_df['mean'] > 0]
```

```python
len(ml_anime_df)

# add in presentation notes that this is a significant drop
# though, does it mean that I should keep those rows b/c they need more visibility?
```

```python
# rename 'mean' to 'score'

ml_anime_df.rename(columns={'mean': 'score'}, inplace=True)
```

```python
ml_anime_df.columns
```

```python
# export dataframe to CSV that will be used for model building
filepath = "MAL_anime_dataset/MAL_data_final.csv"

ml_anime_df.to_csv(filepath, index=False)
```
